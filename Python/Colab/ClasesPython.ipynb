## __1. Generalidades de una clase__

Cuando se habla de una _clase_, se habla directamente de Programación Orientada a Objetos (OOP, por sus siglas en inglés). Las clases son las estructuras con las que se pueden elaborar y crear objetos. Un _objeto_ se define como una instancia de una clase. Las clases se componen de:

* _Métodos:_ funciones que permiten discretizar (subdivir) un problema es una serie de tareas. Emplean variables _locales_ que sólo pueden ser utilizadas dentro del método.

* _Atributos:_ se trata de variables que se consideran _globales_ en la clase. Se pueden utilizar dentro de los métodos.

La estructura _básica_ de una clase es la siguiente:

```
class nombre_clase():
  def metodo1(self, entradas):
    (desarrollo)
    return salidas
  def metodo2(self, entradas):
    (desarrollo)
    return salidas
```

### Ejemplo: Clase calculadora

Se define como _calculadora_ un objeto que permite realizar, por lo menos, las cuatro operaciones básicas: suma, resta, multiplicación y división.

````python
#Estructura general de la calculadora
class calculadora():
  def suma(self, nums):
    s = 0
    for i in nums:
      s += i
    return s
  def resta(self, nums):
    pass
  def division(self, nums):
    pass
  def mult(self, nums):
    pass
````

Con la clase `Calculadora`, puedes crear objetos con los cuales desarrollar las operaciones básicas. Por ejemplo:

````python
#Puedes almacenar la clase en una variable y utilizar sus métodos
c = calculadora()   #La vaiable 'c' es, ahora, un objeto.
c.suma([1,2])
````

### 1.1. Atributos

Los atributos se refieren a las variables consideradas _"globales"_ dentro de una clase. Estas variables pueden ser utilizadas dentro de los métodos de las clases.

````python
class Animal:
  #Atributos
  nombre = ""
  edad = 0
  peso = 0

  def Informacion(self, nombre, edad, peso):
    self.nombre = nombre
    self.edad = edad
    self.peso = peso

#---Fin de la clase---

#Crear animal
perro = Animal();

#Definición de atributos
perro.Informacion("Skipper", 10, 12);

print(perro.nombre + " tiene " + str(perro.edad) + " años y pesa " + str(perro.peso) + " kg.");
````

### 1.3. Constructor

El constructor es un _método especial_ de una clase que permite realizar un desarrollo particular apenas se crea el objeto. Emplea la siguiente estructura:

```
  def __init__ (self, entradas):
    (desarrollo)
```

Puedes emplear objetos que sean _"inicializados"_ con información de interés. Por ejemplo: para el desarrollo de una clase de inventarios, sería útil que al momento de crear el objeto, ya tenga identificados los insumos del negocio.



````python
class Animal:
  #Atributos
  nombre = ""
  edad = 0
  peso = 0

  def __init__(self, nombre, edad, peso):
    self.nombre = nombre
    self.edad = edad
    self.peso = peso

#Animales...
perro = Animal("Skipper", 10, 5)
gato = Animal("Lorenzo", 4, 3)

print(perro.nombre + " es un perro y " + gato.nombre + " es un gato.")
````
